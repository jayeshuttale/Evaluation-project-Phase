### Regression 

```python
xr = df.drop(["AveragePrice","region"],axis=1)
xr.head(3)
```

```python
yr = df.iloc[:,0]
yr.head(3)
```

```python
from sklearn.preprocessing import StandardScaler
smt = StandardScaler()

x_new = smt.fit_transform(xr)
xr = pd.DataFrame(x_new,columns=xr.columns)
xr.head(3)
```

#### Feature selection
* Using variance inflation factor

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def vif(x):
    vif = pd.DataFrame()
    vif["ver"] = x.columns
    vif["VIF"] = [variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return vif
```

```python
vif(xr)
```

* droping High vif feature

```python
xr.drop(['Total Volume'],axis=1,inplace=True)
```

```python
vif(xr)
```

```python
xr.drop('Total Bags',axis=1,inplace=True)
```

```python
vif(xr)
```

**spliting data into train and test**

```python
from sklearn.model_selection import train_test_split
xr_train,xr_test,yr_train,yr_test = train_test_split(xr,yr,test_size=0.2,random_state=42)
```

```python
print('xr_train:',xr_train.shape)
print('xr_test',xr_test.shape)
print('yr_train',yr_train.shape)
print('yr_test',yr_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

### LinearRegression

```python
lr = LinearRegression()
lr.fit(xr_train,yr_train)
print(lr.coef_)
print(lr.intercept_)
```

```python
yr_train_pred = lr.predict(xr_train)
yr_test_pred = lr.predict(xr_test)
```

```python
print("<-----------------train data----------------->")
print("r2_score:",r2_score(yr_train,yr_train_pred))
print("mean_absolute_error:",mean_absolute_error(yr_train,yr_train_pred))
print("mean_squared_error:",mean_squared_error(yr_train,yr_train_pred))
print("adj_r2:" , 1 - ( 1-r2_score(yr_train,yr_train_pred) ) * ( len(yr) - 1 ) / ( len(yr) - xr.shape[1] - 1 ),"\n")

print("<-----------------test data----------------->")
print("r2_score:",r2_score(yr_test,yr_test_pred))
print("mean_absolute_error:",mean_absolute_error(yr_test,yr_test_pred))
print("mean_squared_error:",mean_squared_error(yr_test,yr_test_pred))
print("adj_r2:" , 1 - ( 1-r2_score(yr_test,yr_test_pred) ) * ( len(yr) - 1 ) / ( len(yr) - xr.shape[1] - 1 ))
```
